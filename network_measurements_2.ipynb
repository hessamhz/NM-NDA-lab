```python
import pandas as pd
from datetime import datetime


def packet_capture(pcap_file: str) -> pd.DataFrame:
  packet_df = pd.DataFrame(
      columns=[
          "packet_id", 
          "packet_id_rev",
          "sum_id",
          "src_addr", 
          "src_port", 
          "dst_addr", 
          "dst_port",
          "start_time",
          "size"
          ]
      )
  result_df = pd.DataFrame(
      columns=[
          "sum_id",
          "src_addr", 
          "src_port", 
          "dst_addr", 
          "dst_port",
          "start_time", 
          "duration",
          "total_traffic",
          "bit_rate"
      ]
  )
  nest_asyncio.apply()
  cap = pyshark.FileCapture(pcap_file)
  for packet in cap:
    # not using pandas.append since it is deprecated
    # creating a hash for values of packet to make an id  
    packet_id = hash(f"{packet.ip.src}:{packet.tcp.srcport}{packet.ip.dst}:{packet.tcp.dstport}")
    packet_id_rev = hash(f"{packet.ip.dst}:{packet.tcp.dstport}{packet.ip.src}:{packet.tcp.srcport}")
    # time = datetime.strptime(packet.frame_info.time, '%b %d, %Y %H:%M:%S.%f %Z')
    new_row = pd.DataFrame(
        {
          "packet_id": [packet_id],
          "packet_id_rev": [packet_id_rev],
          "sum_id": [packet_id + packet_id_rev],
          "src_addr": [packet.ip.src],
          "src_port": [packet.tcp.srcport],
          "dst_addr": [packet.ip.dst],
          "dst_port": [packet.tcp.dstport],
          "size": [packet.length],
          "start_time": [packet.sniff_timestamp]
        },
      )
    packet_df = pd.concat([packet_df, new_row], ignore_index=True)
  cap.close()

  
  grouped_df = packet_df.groupby("packet_id")
  for packet_id, group in grouped_df:
    total_size = group["size"].sum()
    start_time = group["start_time"].min()
    end_time = group["start_time"].max()
    duration = float(end_time) - float(start_time)
    bit_rate = int(total_size) / duration
    new_row = pd.DataFrame(
        {
          "sum_id": [group.iloc[0].src_addr],
          "src_addr": [group.iloc[0].src_addr],
          "src_port": [group.iloc[0].src_port],
          "dst_addr": [group.iloc[0].dst_addr],
          "dst_port": [group.iloc[0].dst_port],
          "total_size": [total_size],
          "start_time": [start_time],
          "duration": [duration],
          "bit_rate": [bit_rate]
        },
      )
    result_df = pd.concat([result_df, new_row], ignore_index=True)
  return result_df
```

```python
print(packet_capture("tcp_capture.pcap"))
```
